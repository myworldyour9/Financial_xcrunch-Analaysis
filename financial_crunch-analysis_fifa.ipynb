```python
df=pd.read_csv(filepath_or_buffer = "last_step.csv")
```

## 1.Year-wise Loan Amount Stats

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
year_loan = df.groupby('issue_d - year')['loan_amnt'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=year_loan, x='issue_d - year', y='loan_amnt')

plt.title("Year-wise Loan Amount")
plt.xlabel("Year")
plt.ylabel("Total Loan Amount")
plt.xticks(rotation=45)

plt.show()
```

## 2. Grade & Sub-grade Wise Revolving Balance

```python
grade_loan = df.groupby(['grade','sub_grade'])['loan_amnt'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=grade_loan, x='grade', y='loan_amnt', hue='sub_grade')

plt.title("Grade & Sub-grade Wise Loan Amount")
plt.xlabel("Grade")
plt.ylabel("Total Loan Amount")

plt.show()
```

## 3.Total Payment: Verified vs Non-Verified

```python
verification = df.groupby('verification_status')['loan_amnt'].sum().reset_index()

plt.figure(figsize=(6,5))
sns.barplot(data=verification, x='verification_status', y='loan_amnt')

plt.title("Verified vs Non-Verified Loan Amount")
plt.xlabel("Verification Status")
plt.ylabel("Total Loan Amount")

plt.show()
```

## 4. State-wise Loan Status

```python
state_status = df.groupby(['addr_state','loan_status']).size().reset_index(name='loan_count')

plt.figure(figsize=(12,6))
sns.barplot(data=state_status, x='addr_state', y='loan_count', hue='loan_status')

plt.title("State-wise Loan Status")
plt.xlabel("State")
plt.ylabel("Loan Count")
plt.xticks(rotation=90)

plt.show()
```

## 5. Home Ownership vs Loan Distribution

```python
home_loan = df.groupby('home_ownership')['loan_amnt'].sum().reset_index()

plt.figure(figsize=(6,5))
sns.barplot(data=home_loan, x='home_ownership', y='loan_amnt')

plt.title("Home Ownership vs Loan Amount")
plt.xlabel("Home Ownership")
plt.ylabel("Total Loan Amount")

plt.show()
```

## 1.Year-wise Loan Amount Stats

```python
year_loan = df.groupby('issue_d - year')['loan_amnt'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.lineplot(data=year_loan,
             x='issue_d - year',
             y='loan_amnt',
             marker='o')

plt.title("Year-wise Loan Amount Trend")
plt.xlabel("Year")
plt.ylabel("Total Loan Amount")

plt.show()
```

## 2. Grade & Sub-grade Wise Revolving Balance

```python
grade_data = df.pivot_table(
    values='loan_amnt',
    index='grade',
    columns='sub_grade',
    aggfunc='sum'
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

grade_data = df.groupby(['grade','sub_grade'])['loan_amnt'].sum().reset_index()

plt.figure(figsize=(10,6))

sns.barplot(
    data=grade_data,
    x='grade',
    y='loan_amnt',
    hue='sub_grade'
)

plt.title("Grade & Sub-grade Wise Loan Amount Distribution")
plt.xlabel("Grade")
plt.ylabel("Total Loan Amount")

plt.show()
```

## 3.  Total Payment: Verified vs Non-Verified

```python
verification_data = df.groupby('verification_status')['loan_amnt'].sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plt.pie(
    verification_data,
    labels=verification_data.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Loan Amount Distribution: Verified vs Non-Verified")

plt.show()
```

## 4.State-wise & Last Credit Pull Date Loan Status

```python
loan_status_data = df['loan_status'].value_counts().reset_index()
loan_status_data.columns = ['loan_status', 'count']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.barplot(
    data=loan_status_data,
    x='loan_status',
    y='count'
)

plt.title("Loan Status Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Number of Loans")

plt.xticks(rotation=45)

plt.show()
```

## 5.Home Ownership vs Last Payment Date Stats

```python
home_data = df.groupby('home_ownership')['loan_amnt'].sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plt.pie(
    home_data,
    labels=home_data.index,
    autopct='%1.1f%%',
    startangle=90
)

# create donut hole
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Home Ownership Loan Distribution")

plt.show()
```
